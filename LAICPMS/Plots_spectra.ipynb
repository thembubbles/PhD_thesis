```python
# ---------------- plotting Raman data from text files
        
# --- First step is to import the modules that will be used in the script

import os
import glob  #These two modules are used to look into folders in the PC and access the files


import numpy as np # This module is used for data manipulation
import pandas as pd

import matplotlib.pyplot as plt # This module is what we will use for plotting the spectra

# --- set working directory: Choose the folder that contains your spectra text files

data_dir = 'LA-ICP-MS/Data/spectra_files/'
```

```python
# --- look into working directory and create a list with selected files - .csv
   
os.chdir(data_dir) 

spectra_files = glob.glob("*.csv") #loop through folder (** looks into subfolders) and get .csv files
    
print(spectra_files)
print(len(spectra_files))
```
